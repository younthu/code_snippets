Extract keys from json files under folder `regions/`

The json structure is something like:
~~~json
{
    "SupportedRegions": {
        "e3171dd9-9a5f-e5be-b36c-cc7c4f3f3bcf": [
            {
                "name": "australiaeast",
                "geo": "AsiaPacific",
                "regionGroup": "Australia",
                "isHeroRegion": true,
                "allowTenantsPolicy": "None",
                "allowedSubscriptions": [],
                "status": "Initial",
                "partners": [
                    "e3171dd9-9a5f-e5be-b36c-ccxxxxxxxxxxx"
                ]
            },
            {
                "name": "canadacentral",
                "geo": "Americas",
                "regionGroup": "Canada",
                "isHeroRegion": false,
                "allowTenantsPolicy": "None",
                "allowedSubscriptions": [],
                "status": "Initial",
                "partners": [
                    "e3171dd9-9a5f-e5be-b36c-ccxxxxxxxxxx"
                ]
            },
        ]
    }
}
~~~

The purpose is to get all those "name".

```python
import json

files = ["regions/fairfax/SupportedRegions_Win365.json",
         "regions/public/SupportedRegions_PowerAutomate.json",
         "regions/public/int/SupportedRegions_Fidalgo.json",
         "regions/public/int/SupportedRegions_PowerAutomate.json",
         "regions/public/int/SupportedRegions_Win365.json",
         "regions/public/ppe/SupportedRegions_Fidalgo.json",
         "regions/public/ppe/SupportedRegions_Win365.json",
         "regions/public/prod/SupportedRegions_Fidalgo.json",
         "regions/public/prod/SupportedRegions_Win365.json",
         "regions/public/prod/prna01/SupportedRegions_Win365.json",
         "regions/public/prod/prna02/SupportedRegions_Win365.json",
         "regions/public/sh/SupportedRegions_Fidalgo.json",
         "regions/public/sh/SupportedRegions_Win365.json"
        ];
```

```python
region_list = []
for json_file in files:
    with open(json_file, "r") as config:
        j = json.load(config)
        regions = j["SupportedRegions"]
        
        for sub in regions:
            for item in regions[sub]:
                print(item["name"])
                region_list.append(item["name"])
```

```python
uni_list = list(set(region_list))
len(region_list),len(uni_list)
```
